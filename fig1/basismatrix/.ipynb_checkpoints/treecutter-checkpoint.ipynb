```python
# basics
import numpy as np
import pandas as pd
import scanpy as sc
import os


# viz
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# dendrogram
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import cut_tree


#For editable text. Except latex text is still shapes sadly
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

font = {'size' : 6}
lines = {'linewidth' : 0.5}
fig = {'figsize' : (2.5, 1.5)}
mpl.rc('font', **font)
mpl.rc('lines', **lines)
mpl.rc('figure', **fig)
#Set style
sns.set(style="whitegrid", palette="pastel", color_codes=True)

from matplotlib.colors import ListedColormap

from scipy.stats import mannwhitneyu

np.random.seed(31)
```

```python
def save_figure_pdf(fig, path_to_save):
	'''
	Utility fxn to save a figure without having to retype all the required options.
	Input:
		fig - a matplotlib.pyplot.figure instance with the figure you'd like to save
		path_to_save - the path you'd like to save the figure to
	Returns: Nothing but there should now be a pdf version in 300 dpi with a transparent bkg at path_to_save
	'''
	fig.savefig(path_to_save, dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
# function to backtrack the ordering of the labels
# since we don't know the original object from python
def backtrackLabels(dn):
    leaves = dn["dendrogram_info"]["leaves"]
    ogLabels = ["" for i in range(len(leaves))]
    ivlLabels = dn["dendrogram_info"]['ivl'] # the output

    ivlIndex = 0
    for leaf in leaves:
        ogLabels[leaf] = ivlLabels[ivlIndex]
        ivlIndex += 1 # increment to move to the next element
    return(ogLabels)
```

```python
# function to backtrack the ordering of the labels
# since we don't know the original object from python

# adjust for tabula sapiens since the adata object is large to rip accordingly
def backtrackLabels_TSP(leafDF, ivlDF):
    leaves = leafDF.index.tolist()
    ogLabels = ["" for i in range(len(leaves))]
    ivlLabels = ivlDF.index.tolist() # the output

    ivlIndex = 0
    for leaf in leaves:
        ogLabels[leaf] = ivlLabels[ivlIndex]
        ivlIndex += 1 # increment to move to the next element
    return(ogLabels)
```

```python
compartments = ["stromal", "epithelial"]
```

```python
import os
```

```python
tspDendoPath = os.getcwd() + "/"
```

```python
figSavePath = ""
```

```python
mergedLeavesAcrossCompartments = pd.DataFrame()

for compartment in compartments:
    tspLeaves = pd.read_csv(tspDendoPath + "tsp_leafNames_" + compartment + "-_cleanTSP_07312021.csv", index_col = 0)

    tspIVL = pd.read_csv(tspDendoPath + "tsp_ivl_" + compartment + "-_cleanTSP_07312021.csv", index_col = 0)

    tspLinkage = pd.read_csv(tspDendoPath + "tsp_linkageDF_" + compartment + "-_cleanTSP_07312021.csv",
                            index_col = 0)

    tspDcoord = pd.read_csv(tspDendoPath + "tsp_dcoord_" + compartment + "-_cleanTSP_07312021.csv",
                           index_col = 0)
    
    heights = tspDcoord.iloc[:,1]
    
    if compartment == "stromal" or compartment == "endothelial":
        cutHeight = np.max(heights) * .1
    elif compartment == "epithelial":
        cutHeight = np.max(heights) * .2
 
    print("COMPARTMENT = ", compartment, " MAX HEIGHT: ", np.max(heights))
    
    plt.figure(figsize = (20, 20))
    tspZ = tspLinkage.values
    tspLabelList = backtrackLabels_TSP(tspLeaves, tspIVL)
    hierarchy.set_link_color_palette(["#1b4965"])
    
    dn = dendrogram(tspZ, color_threshold = 0,
                    labels = tspLabelList, 
                    link_color_func=lambda k: "#1b4965",
                    
                    leaf_font_size = 8)
    plt.xticks(rotation = "90")
    
    # this is how you add the cut line
    # code for this line: 
    # https://www.python-graph-gallery.com/401-customised-dendrogram
    plt.axhline(y = cutHeight, c = 'grey', lw = 1, linestyle = 'dashed') 
    plt.gcf()
    
    fname = figSavePath + compartment + "_cleanTSP_sigMatv18_noGROUPING_09102021.pdf"
    save_figure_pdf(plt, fname)
    
    cut = cut_tree(tspZ, height = cutHeight)
    cutDF = pd.DataFrame(data = cut, index = tspLabelList)
    cutDF.columns = ["cluster"]
    
    # now merge the labels on the basis of cut tree
    allRows = []
    clusters = np.unique(cutDF["cluster"])
    for clus in clusters:
        thisClus = cutDF[cutDF["cluster"] == clus]

        joined = ""
        for i in list(thisClus.index):
            joined += i + "/"

        joined = joined[:-1] # remove the last "/"

        allRows += [joined]


    mergedLabels = pd.DataFrame(index = allRows, data = clusters)
    mergedLabels.columns = ["clusters"]
    
    # update the df holding them all
    mergedLeavesAcrossCompartments = pd.concat([mergedLeavesAcrossCompartments, mergedLabels])
    mergedLabels.to_csv(figSavePath + compartment + "_clusters_cleanTSP_sigMatv18_09102021.csv")
   
```

```python
mergedLeavesAcrossCompartments.shape
```

```python
# keys are the parent clusters
# vals are the individual cell types in the clusters
clusterDict = {}
for i in mergedLeavesAcrossCompartments.index.tolist():
    clusterDict[i] = list(i.split("/"))
```

```python
# make a reverse dictionary where the keys are the original labels
# and the values are their super cluster
revDict = {}
for superClus in clusterDict:
    cellsThisSuperClus = clusterDict[superClus]
    
    for cell in cellsThisSuperClus:

        revDict[cell] = superClus
```

```python
lumpDF = pd.DataFrame.from_dict(revDict, orient = "index")
lumpDF.columns = ["joinedLabel"]
```

```python
# now add the immune compartment in

# all TSP metadata
meta = pd.read_csv("/Users/kayaneh/Documents/deconvolution/revision1/07262021_cleanTSP_sigmat11/cleanTSP_obsDF.csv", index_col = 0)
cell, cts = np.unique(meta["cell_ontology_class"], return_counts = True)

toExclude = []

# exclude granulocyte because we have basophil/neutrophil
# exclude t follicular helper cell since it's a t cell subtype
# exclude myeloid cell since it's only in pancreas/prostate annotations
# exclude 'mesenchymal stem cell' since this isn't in the immune compartment
badAnnot = ["leucocyte", "immune cell", "granulocyte", "erythroid lineage cell",
           't follicular helper cell', "myeloid cell", 'mesenchymal stem cell']

# t cell
tcellExclude = [i for i in cell if ("-t cell" in i or " t cell" in i) and "nk" not in i]
print(tcellExclude)

# nk t cell exclude/nk cell subtype
nkTExclude = [i for i in cell if "natural" in i or "nk" in i and i != "nk cell"]

monocyteExclude = [i for i in cell if "monocyte" in i and i != "monocyte"]
monocyteExclude

neutExclude = [i for i in cell if "neutrophil" in i and i != "neutrophil"]
print(neutExclude)

thymoExclude = [i for i in cell if "thymocyte" in i and i != "thymocyte"]
thymoExclude

dendoExclude = [i for i in cell if "dendritic" in i and i != 'mature conventional dendritic cell'] + ["langerhans cell"]
print(dendoExclude)

bcellExcl = [i for i in cell if " b cell" in i and i != "b cell"]
print(bcellExcl)

allExclusions =  bcellExcl + badAnnot + tcellExclude + nkTExclude + monocyteExclude + neutExclude + thymoExclude + dendoExclude

len(allExclusions)

immuneCells = np.unique(meta[meta["compartment"] == "immune"]["cell_ontology_class"])

len(immuneCells)

# these are the cells to work with
immuneCompartment = np.setdiff1d(immuneCells, allExclusions).tolist()
print(immuneCompartment)

immuneCompDF = pd.DataFrame(index = immuneCompartment, 
             data = immuneCompartment,
            columns = ["joinedLabel"])

# merge the erythroid progenitor and the erythrocyte
immuneCompDF.loc["erythroid progenitor"] = "erythrocyte/erythroid progenitor"
immuneCompDF.loc["erythrocyte"] = "erythrocyte/erythroid progenitor"

# add immune to lumped
allCompDF = pd.concat([lumpDF, immuneCompDF])

# size of the column space
len(np.unique(allCompDF["joinedLabel"])) + len(immuneCompDF)
```
