```python
savePath = ""
sourceDataPath = "/Users/kayaneh/Documents/deconvolution/aip/source_data/suppFig1_cfRNAQC_sourceData.xlsx"
figEnd = "_08242021.pdf"
```

```python
# these are the 95th percentile cutoffs determined by mira

# from mira
ribo = 0.2 
deg = 0.4
intron = 3
```

```python
cmap = ListedColormap(['0.25', 'red'])
```

```python
def save_figure_pdf(fig, path_to_save):
	'''
	Utility fxn to save a figure without having to retype all the required options.
	Input:
		fig - a matplotlib.pyplot.figure instance with the figure you'd like to save
		path_to_save - the path you'd like to save the figure to
	Returns: Nothing but there should now be a pdf version in 300 dpi with a transparent bkg at path_to_save
	'''
	fig.savefig(path_to_save, dpi = 300, bbox_inches = 'tight', transparent = True)
```

```python
nafldBase = "/Users/kayaneh/Documents/deconvolution/nafld/remapped_unstranded/"
intronPath = nafldBase + "intron_exon_ratios_nafld_unstrandedTS3.txt"
riboPath = nafldBase + "nafld_ribo_frac_unstrandedTS3.txt"
prime3Path = nafldBase + "deg_3prime_bias_frac_1_nafld_unstrandedTS3.txt"
```

```python
intronDistrib = pd.read_csv(intronPath, sep = "\t", index_col = 0)

riboDistrib = pd.read_csv(riboPath, sep = ",", index_col = 0)

prime3Distrib = pd.read_csv(prime3Path, sep = "\t",index_col = 0)
```

```python
allQCDat = intronDistrib.join(riboDistrib)
allQCDat = allQCDat.join(prime3Distrib)
```

```python
with pd.ExcelWriter(sourceDataPath, mode = 'a') as writer:
            allQCDat.to_excel(writer, sheet_name = 'ChalasaniEtAl', engine = 'xlsxwriter')
```

```python
plt.figure(figsize = (5, 5))

sns.boxplot(data = intronDistrib, color = "#cae9ff", fliersize = 0)


ax = sns.swarmplot(data = intronDistrib, color = "0.25", size = 3)

for col in ax.collections:
    y = col.get_offsets()[:,1]   
    perc = [intron]
    col.set_cmap(cmap)
    col.set_array(np.digitize(y, perc))
plt.title("Intron/Exon Ratio")
plt.ylabel("Ratio(Intron / Exon)")
#save_figure_pdf(plt, savePath + "nafld_intronExonQC" + figEnd)
```

```python
plt.figure(figsize = (5, 5))
sns.boxplot(data = riboDistrib, color = "#cae9ff", fliersize = 0)
ax = sns.swarmplot(data = riboDistrib, color = "0.25", size = 3)
for col in ax.collections:
    y = col.get_offsets()[:,1]
    perc = [ribo]
    col.set_cmap(cmap)
    col.set_array(np.digitize(y, perc))
plt.title("Fraction Ribosomal Reads")
plt.ylabel("Ribosomal Read Fraction")
save_figure_pdf(plt, savePath + "nalfd_riboFracQC" + figEnd)
```

```python
plt.figure(figsize = (5, 5))
sns.boxplot(data = prime3Distrib, color =  "#cae9ff", fliersize = 0)
ax = sns.swarmplot(data = prime3Distrib, color = "0.25", size = 3)
for col in ax.collections:
    y = col.get_offsets()[:,1]
    perc = [deg]
    col.set_cmap(cmap)
    col.set_array(np.digitize(y, perc))
plt.title("3' Bias Fraction")
plt.ylabel("3' Bias Fraction")
save_figure_pdf(plt, savePath + "nafld_3PrimeBiasQC" + figEnd)
```
