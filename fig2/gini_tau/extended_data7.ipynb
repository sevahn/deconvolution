# plotting preliminaries

```python
def save_figure_pdf(fig, path_to_save):
	'''
	Utility fxn to save a figure without having to retype all the required options.
	Input:
		fig - a matplotlib.pyplot.figure instance with the figure you'd like to save
		path_to_save - the path you'd like to save the figure to
	Returns: Nothing but there should now be a pdf version in 300 dpi with a transparent bkg at path_to_save
	'''
	fig.savefig(path_to_save, dpi = 300,
                bbox_inches = 'tight', transparent = True)
```

```python
pal1 = {"Amaranth M P":"893168","Dark Purple":"4a1942",
        "Dark Purple 2":"1b4965","Rich Black FOGRA 39":"050404","Platinum":"eaeaea"}
pal2 = {"Light Cornflower Blue":"8ecae6","Blue Green":"219ebc",
 "Prussian Blue":"023047","Orange":"fb8500", "Honey Yellow":"ffb703"}
```

```python
def hexCoolorsPal(coolPal, keyList):
    for i in coolPal:
        coolPal[i] =  "#" + coolPal[i]
    
    final_pal = {}
    for i, cell in enumerate(keyList):
        final_pal[cell] = list(coolPal.values())[i]
    return(final_pal)

```

```python
brain = pd.read_csv("brain_gini_tau_02142021.csv", sep = ",", index_col = 0)
liver = pd.read_csv("liver_gini_tau_02142021.csv", sep = ",", index_col = 0)
kidney = pd.read_csv("kidney_gini_tau_02142021.csv", sep = ",", index_col = 0)
placenta = pd.read_csv("placenta_gini_tau_02142021.csv", sep = ",", index_col = 0)
testis = pd.read_csv("testis_gini_tau_02142021.csv", sep = ",", index_col = 0)
heart = pd.read_csv("heart_gini_tau_04082021.csv", sep = ",", index_col = 0)
```

```python
spec_pal = hexCoolorsPal(pal1,np.unique(brain["RNA tissue specificity"]) )
distrib_pal = hexCoolorsPal(pal2, np.unique(brain["RNA tissue distribution"]))
```

```python
brain["tiss"] = "brain"
brain.columns = ["tau", "gini"] + brain.columns.tolist()[2:]

liver["tiss"] = "liver"
liver.columns = ["tau", "gini"] + liver.columns.tolist()[2:]

kidney["tiss"] = "kidney"
kidney.columns = ["tau", "gini"] + kidney.columns.tolist()[2:]

placenta["tiss"] = "placenta"
placenta.columns = ["tau", "gini"] + placenta.columns.tolist()[2:]

testis["tiss"] = "testis"
testis.columns = ["tau", "gini"] + testis.columns.tolist()[2:]

heart["tiss"] = "heart"
heart.columns = ["tau", "gini"] + heart.columns.tolist()[2:]
```

```python
allTiss = pd.concat([brain, liver, kidney, placenta, testis], axis = 0)
allTiss = pd.concat([brain, liver, kidney, placenta, heart], axis = 0)
```

```python
allTissDistrib = allTiss.drop("RNA tissue specificity", axis = 1)
allTisSpecificity = allTiss.drop("RNA tissue distribution", axis = 1)
```

```python
meltedTiss_distrib = pd.melt(allTissDistrib, 
                             id_vars = ["RNA tissue distribution", "tiss"],
                             value_vars = ["gini", "tau"])

meltedTiss_specific = pd.melt(allTisSpecificity, 
                             id_vars = ["RNA tissue specificity", "tiss"],
                             value_vars = ["gini", "tau"])
```

```python
plt.figure(figsize = (6, 6))

g = sns.displot(meltedTiss_specific,
            x = "value", 
        multiple = "stack",
        kind = "kde",
        bw_adjust=.3,
        hue = "RNA tissue specificity",
        common_norm = False,
        palette = spec_pal,
        col = "variable",
        row = "tiss",
        facet_kws = {"sharey":False,
                    "sharex": False})

save_figure_pdf(plt, figSavePath +  "ginitau_rnaSpec_allTiss" + figEnd)



```

```python
plt.figure(figsize = (6, 6))
sns.displot(meltedTiss_distrib,
            x = "value", 
        multiple = "stack",
        kind = "kde",
        bw_adjust=.3,
        hue = "RNA tissue distribution",
        common_norm = False,
        palette = distrib_pal,
        col = "variable",
           row = "tiss",
            hue_order = ["Detected in single",
                        "Detected in some",
                        "Detected in many",
                        "Detected in all"],
            facet_kws = {"sharey":False,
                    "sharex": False})

save_figure_pdf(plt, figSavePath + "ginitau_rnaDistrib_allTiss_" + figEnd)
```
